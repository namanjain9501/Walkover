program to print even length words in a string

```python
def printWords(s):
    s = s.split(' ') 
    for word in s: 
        if len(word)%2==0:
            print(word) 

s=input("Enter string=")
printWords(s)            
```

Program to accept the strings which contains all vowels

```python
def check(string) :

	string = string.lower()
	vowels = set("aeiou")
	s = set({})
	for char in string :
		if char in vowels :
			s.add(char)
		else:
			pass
	
	if len(s) == len(vowels) :
		print("Accepted")
	else :
		print("Not Accepted")

if __name__ == "__main__" :
	
	string=input("Enter string=")

	
	check(string)
```

Find words which are greater than given length k

```python
# function find string greater than length k
def string_k(k, str):
	string = []
	text = str.split(" ")
	
	for x in text:
		if len(x) > k:
			string.append(x)
	return string

k=int(input("Enter k="))	
str=input("Enter string")
print(string_k(k, str))
```

program to split and join a string

```python
def split_string(string):
    list_string = string.split(' ')
    return list_string

def join_string(list_string):
    string = '-'.join(list_string)
    return string

string=input("Enter string=")
# Splitting a string
list_string = split_string(string)
print("After Splitting: ",list_string)

# Join list of strings
res_string = join_string(list_string)
print("After joining: ",res_string)
```

Maximum frequency character in String

```python
from collections import Counter

string=input("Enter string=")
print(string)

result= Counter(string)
result= max(result, key=result.get)

print("Most frequent character: ",result)
```

Multiply all numbers in the list

```python
def multiplyList(myList) :
    result = 1
    for x in myList:
         result = result * x
    return result
     

list1 = [2, 4, 6]
list2 = [3, 2, 4]
print(multiplyList(list1))
print(multiplyList(list2))
```

```python
def swapList(newList):
	size = len(newList)
	
	# Swapping
	temp = newList[0]
	newList[0] = newList[size - 1]
	newList[size - 1] = temp
	
	return newList

newList = [45,34,66,7,8,21]

print(swapList(newList))
```

least frequent character in a string

```python
string=input("Enter string=")
print(string)
char_freq={}
for i in string:
    if i in char_freq:
        char_freq[i]=char_freq[i]+1
    else:
        char_freq[i] = 1
result= min(char_freq, key = char_freq.get)
print("Least frequent character: ",result)
```

Check if a given string is binary string or not

```python
string = input("Enter string=")
b = {'0','1'}
t = set(string)

if b == t or t == {'0'} or t == {'1'}:
  print("String is a binary string.")
else:
  print("String is not a binary string.")
```

inverse pyramid 

```python
row = int(input('Enter number of rows required: '))

for i in range(row,0,-1):
    for j in range(row-i):
        print(' ', end='') 
    
    for j in range(2*i-1):
        print('*',end='') 
    print() # printing new line
```

to sort list of tuples by second item

```python
def Sort_Tuple(tup):
	lst = len(tup)
	for i in range(0, lst):
		
		for j in range(0, lst-i-1):
			if (tup[j][1] > tup[j + 1][1]):
				temp = tup[j]
				tup[j]= tup[j + 1]
				tup[j + 1]= temp
	return tup

tup =[('an', 24), ('am', 10), ('Intern', 28),
	('I', 5), ('as', 20), ('working', 15)]
		
print(Sort_Tuple(tup))
```
